```python
#環境の定義

__credits__ = ["Kallinteris-Andreas"]

import os
import warnings
import time
import matplotlib.pyplot as plt
import mujoco
import mujoco.viewer
import glfw
import math
import numpy as np
import pytest
import random
from gymnasium import utils
from gymnasium.envs.mujoco import MujocoEnv
from gymnasium.error import Error
from gymnasium.spaces import Box
from gymnasium import spaces
from gymnasium.utils.env_checker import check_env
from scipy.spatial.transform import Rotation as R
from collections import deque




class Mouse1massEnv(MujocoEnv, utils.EzPickle):
    """
    Gymansium.MujocoEnv`環境APIを使った，マイクロマウスの強化学習環境\n
    衝突回避と前進行動を両立させるタスク．\n
    mujoco envを継承しています．
    """

    metadata = {
        "render_modes": [
            "human",
            "rgb_array",
            "depth_array",
        ],
    }

    def __init__(self, xml_file="my_xmls/16x16maze_model_NO_merikomi_masatu.xml", frame_skip=2, **kwargs):
        utils.EzPickle.__init__(self, xml_file, frame_skip, **kwargs)

        MujocoEnv.__init__(
            self,
            xml_file,
            frame_skip=frame_skip,
            observation_space=None,  # needs to be defined after
            default_camera_config={},
            camera_name="camera_name",
            **kwargs,
        )
        

        self.metadata = { #またメタデータがある。インスタンス変数？
            "render_modes": [
                "human",
                "rgb_array",
                "depth_array",
            ],
            "render_fps": int(np.round(1.0 / self.dt)),
        }

        obs_size = 17 #self.data.qpos.size + self.data.qvel.size 
        #たとえば，観測空間に位置と速度を入れたいのであれば，その数をここに入れる。次元数なので数え方は０始まりではない。
        # self.wall_hit = 0
        #マイクロマウスのパラメータ用変数の用意
        self.reward_graph = []
        self.deg_difference_graph = []
        self.x_velocity_reward_graph = []
        self.odom_vel_graph = []
        self.epi_reward_graph = []
        self.x_veloci_reward_graph = []

        self.kakudo_graph = []
        self.god_kakudo_graph = []
        self.god_x_veloci_graph = []
        self.god_y_veloci_graph = []
        self.god_z_veloci_graph = []
        self.x_veloci_graph = []
        self.y_veloci_graph = []
        self.z_veloci_graph = []
        self.god_x_pos_graph = []
        self.god_y_pos_graph = []
        self.god_z_pos_graph = []
        self.x_pos_graph = []
        self.y_pos_graph = []
        self.z_pos_graph = []
        self.gyro_x_graph = []
        self.gyro_y_graph = []
        self.gyro_z_graph = []
        self.acc_x_graph = []
        self.acc_y_graph = []
        self.acc_z_graph = []
        self.mouse_xpos_graph = []
        self.mouse_ypos_graph = []
        self.lastxpos_graph = []
        self.lastypos_graph = []
        self.odom_xpos_graph = []
        self.odom_ypos_graph = []
        self.god_xpos_graph = []
        self.god_ypos_graph = []
        self.odom_mouse_angle_rad = math.pi/2
        self.odom_xpos = 0.27
        self.odom_ypos = 0.18
        self.sec_0_flag = False
        self.sec_1_flag = False
        self.sec_2_flag = False
        self.sec_3_flag = False
        self.sec_4_flag = False
        self.sec_5_flag = False
        self.sec_6_flag = False
        self.sec_7_flag = False
        self.sec_8_flag = False
        self.sec_9_flag = False
        self.sec_10_flag = False
        self.sec_11_flag = False
        self.sec_12_flag = False
        self.sec_13_flag = False
        self.goal_random = 8 #適当に、0と1以外の値にした
        self.wall_mode = 0
        self.mouse_angle_rad = math.pi/2
        self.leftwall = 1
        self.rightwall = 1
        self.forwardwall = 1
        self.backwall = 1
        self.mouse_xpos = 0
        self.mouse_ypos = 0
        self.mouse_vel = 0
        self.now_time = 0
        self.prev_time = 0
        self.delta_t = 0.01
        self.x_veloci = 0
        self.y_veloci = 0
        self.z_veloci = 0
        self.x_pos = 0
        self.y_pos = 0
        self.z_pos = 0
        self.gear = 9.0e-3
        self.wheel_r = 0.0135
        self.tread = 0.072
        self.right_rotation_sum = 0
        self.left_rotation_sum = 0
        self.encoder_pulses_per_turn = 4096
        self.odom_x_graph = []
        self.odom_y_graph = []
        self.stepcount = 0
        self.stepcount_for_resetgraph = 0
        self.angsum_obs = 0
        self.epi_reward = 0
        self.reward_average_graph = []
        self.epi_steplength_graph = []
        self.valiance = []
        self.valiance_average_graph = []
        self.obsLS_graph = []
        self.obsLF_graph = []
        self.obsRS_graph = []
        self.obsRF_graph = []
        self.range_reward_x_graph = []
        self.range_reward_y_graph = []
        self.xpos_graph = []
        self.ypos_graph = []
        self.firstxyzahyou = self.init_qpos[0:3]
        self.firstquat = self.init_qpos[3:7]
        self.first_ang = (0, 0, 90)
        self.first_ang_graph = []
        self.goal_pos_number = 2
        self.geom_id1 = mujoco.mj_name2id(self.model, mujoco.mjtObj.mjOBJ_GEOM, "right_wheel")  # ジョイント名からIDを取得
        self.geom_id2 = mujoco.mj_name2id(self.model, mujoco.mjtObj.mjOBJ_GEOM, "left_wheel")  # ジョイント名からIDを取得

        self.mouse_body_id = mujoco.mj_name2id(self.model, mujoco.mjtObj.mjOBJ_BODY, "mouse_body")  # ボディ名からIDを取得
        self.mouse_position = 0

        # 過去のセンサ値を保存するリスト
        self.acc_xsensor_history = [0, 0, 0]
        self.acc_ysensor_history = [0, 0, 0]
        self.gyro_zsensor_history = [0, 0, 0]
        self.LSsensor_history = [0, 0, 0]
        self.LFsensor_history = [0, 0, 0]
        self.RFsensor_history = [0, 0, 0]
        self.RSsensor_history = [0, 0, 0]

        # 座標分散報酬計算用キューの最大サイズを5に設定
        self.zahyou_deque = deque(maxlen=50)
        # 足跡観測用キューの最大サイズを5に設定
        # self.asiato_deque = deque(maxlen=200)
        # 壁めり込んだことを検出する用のキュー
        self.force_sensor_deque = deque(maxlen=15)
        # 壁ぶつかった後はしばらくは後退したら報酬を与えるための、速度のログdeque
        self.hit_deque = deque(maxlen=25)

        # 各観測用deque
        self.LS_sensor_deque = deque([0, 0, 0, ], maxlen=3)
        self.LF_sensor_deque = deque([0, 0, 0, ], maxlen=3)
        self.RF_sensor_deque = deque([0, 0, 0, ], maxlen=3)
        self.RS_sensor_deque = deque([0, 0, 0, ], maxlen=3)
        self.act_Left_deque = deque([0], maxlen=1)
        self.act_Right_deque = deque([0], maxlen=1)
        self.veloci_sensor_deque = deque([0], maxlen=1)
        self.acc_x_sensor_deque = deque([0], maxlen=1)
        self.acc_y_sensor_deque = deque([0, 0, 0, 0, 0], maxlen=5)
        self.gyro_z_sensor_deque = deque([0, ], maxlen=1)
                
        self.action = [0, 0]



        self.observation_space = Box( #観測空間を定義．gymnasium.sapcesのBoxを使用することで簡単に定義できる．Boxは連続値を扱う．
            low=-np.inf, high=np.inf, shape=(obs_size,), dtype=np.float64 #float64型の，-無限から無限までの値を取る，11次元の観測空間を用意
        )

    def get_now_wheel_ang(self, model, data): #今のタイヤの回転角度と速度（角度と角速度）を取得．
        """
            現在のタイヤの回転角度と速度（と角速度）を取得する．\\
            data.actuator().length,velocityで取得できる値はactuatorのgear倍されているので，それで割ることによって打ち消す，
            Returns:
                odm_right = 右タイヤの回転角度[rad]（基準からの総量） \\
                odm_left = 左タイヤの回転角度[rad]（基準からの総量） \\
                (vel_left) = 左タイヤの回転角速度 (単位時間あたり) \\
            """
        right_wheel_ang = data.actuator('right').length[0]/self.gear # 得た値をgearで割ったもの＝ギアの付いた先の、タイヤ自体の回転した角度(rad)
        left_wheel_ang = data.actuator('left').length[0]/self.gear # タイヤの角度
        vel_left = data.actuator('left').velocity[0]/self.gear # ＝タイヤの回転角速度
        return right_wheel_ang, left_wheel_ang # タイヤの角度を取得
    

    def get_pulse_count(self, pre_right_wheel_ang, pre_left_wheel_ang, right_wheel_ang, left_wheel_ang): #回転量をパルスにする エンコーダの役割をする関数
        """
            回転量の差から，出たパルスの数を求める． (エンコーダの役割をする関数)\\
            
            Returns:
                right_pulse_num = 右タイヤのパルス数 \\
                left_pulse_num = 左タイヤのパルス数
            """

        pre_right_rotation_sum = self.right_rotation_sum #１つまえの合計回転量取得
        pre_left_rotation_sum = self.left_rotation_sum
        self.right_rotation_sum += (right_wheel_ang - pre_right_wheel_ang)/(2*math.pi) *4096 #1周あたり4096段階で，細かい回転量を計測（分解能ほぼ無限のエンコーダ）（小数で、0から1で回転を表していたものを0から4096で表すようにした）
        self.left_rotation_sum += (left_wheel_ang - pre_left_wheel_ang)/(2*math.pi) *4096 #1周あたり4096段階で，細かい回転量を計測（分解能ほぼ無限のエンコーダ）
        right_pulse_num = int(self.right_rotation_sum) - int(pre_right_rotation_sum) #現在の回転量を足した後の回転合計値ー足す前の回転合計値　これのintが１以上なら、１パルス出たことになる。パルスが出なくても回転量の合計は溜まっていく
        left_pulse_num = int(self.left_rotation_sum) - int(pre_left_rotation_sum)
        # print("回転量をパルス段階に変換　この値の差が，実際に出たパルスとなる：", right_pulse_num)
        return right_pulse_num, left_pulse_num #出たパルスの数

    def calc_odometry_at_pulse(self, right_pulse_num, left_pulse_num): #今のタイヤの回転角度と速度（角度と角速度）を求める．そのご、タイヤの回転による移動距離と移動速度を求め、車体の速度と角速度を求める。
        """
            パルス数を元に、ロボットの速度と回転角速度を求める．（オドメトリによる推定を行う）\\
            Returns:
                mouse_vel = ロボットの推定速度 \\
                left_pulse_num = ロボットの推定回転速度
            """
        right_wheel_move = 2*math.pi * self.wheel_r * (right_pulse_num/self.encoder_pulses_per_turn) #２πｒ×回転量（発生パルス数÷1周のパルス数）で、タイヤの移動距離を求める．
        left_wheel_move = 2*math.pi * self.wheel_r * (left_pulse_num/self.encoder_pulses_per_turn) #２πｒ×回転数で移動距離を求める．
        right_wheel_vel = right_wheel_move/self.delta_t #移動距離を時間で割れば、単位時間あたりの移動距離、つまり速度が求められる
        left_wheel_vel = left_wheel_move/self.delta_t #移動距離から，速度を求める
        robot_vel = (right_wheel_vel + left_wheel_vel)/2 #左右のタイヤの速度の平均が、車体全体の速度
        robot_radvel = (right_wheel_vel - left_wheel_vel)/self.tread #これがマウスの旋回角速度（rad/s）左右のタイヤの速度の差をtread幅の長さで割ったものが、角速度になるため。
        return robot_vel, robot_radvel #これで，エンコーダによる各タイヤの推定速度が求められた
    
    def friction_randam_change(self):
        random_number_right = np.random.rand()
        random_number_left = np.random.rand()
        # 右タイヤの摩擦係数をランダムに変更
        if random_number_right < 0.05: #5%の確率で摩擦係数を0にする
            self.model.geom_friction[self.geom_id1][0] = 0
        else:
            self.model.geom_friction[self.geom_id1][0] = 1

        # 左タイヤの摩擦係数をランダムに変更
        if random_number_left < 0.05:
            self.model.geom_friction[self.geom_id2][0] = 0
        else:
            self.model.geom_friction[self.geom_id2][0] = 1

    def step(self, action):
        """
          シミュレーションを1step進める。 \\
          Returns: \\
            observation, 観測空間。その行動後のエージェントが見えている状態を返す \\
            reward, 報酬の合計値 \\
            terminated = False, タスクの早期達成または強制終了条件にひっかかったわけではないが、意味のない行動をし続ける場合に陥ったときなどにエピソードを終了するフラグ。本環境ではそういう場面はないため常にFalseとなる。 \\
            truncated = タスクの早期達成または強制終了が起きたときにそれを示すもの。壁にぶつかったら強制終了が起きる。ぶつかったとき以外はFalse。 \\
            info = （現状未確認。何かの情報が入っている？）\\
        """
        #変数定義
        # print("step")
        terminated = False
        truncated = False
        goal_reward = 0
        died_reward = 0
        survive_reward = 0.001 # より速くゴールに到達するために、ロスタイム報酬
        body_id = mujoco.mj_name2id(self.model, mujoco.mjtObj.mjOBJ_BODY, "mouse_body")  # ボディ名からIDを取得
        self.act_Left_deque.append(action[0])
        self.act_Right_deque.append(action[1])
        # self.action = action

        ####オドメトリ処理開始##################################################################################
        self.prev_time = self.data.time #シミュレーション進める前の時間を取得
        pre_right_wheel_ang, pre_left_wheel_ang = self.get_now_wheel_ang(self.model, self.data) #今のタイヤの回転角度と速度（角度と角速度）を取得．
        self.friction_randam_change() #摩擦係数をランダムに変更
        self.do_simulation(action, self.frame_skip)####シミュレーション実行###########################################
        self.stepcount += 1

        self.now_time = self.data.time #1step進んだ後のシミュレーションの時間を取得
        self.delta_t = self.now_time - self.prev_time #シミュレーション1ステップにかかった時間を取得
        self.prev_time = self.now_time #シミュレーション前の時間の変数を更新
        right_wheel_ang, left_wheel_ang = self.get_now_wheel_ang(self.model, self.data)# エンコーダをもう一度読み取る　（行動後の値を取得）
        
        right_pulse_num, left_pulse_num = self.get_pulse_count(pre_right_wheel_ang, pre_left_wheel_ang, right_wheel_ang, left_wheel_ang) #シミュレーション前後のエンコーダ値から，この一瞬に出たパルス数を算出し，
        odom_mouse_vel, odom_mouse_radvel = self.calc_odometry_at_pulse(right_pulse_num, left_pulse_num) # パルス数から，移動速度と角速度の推定値を求める。　これで，エンコーダによるマウスの速度・角速度が推定できた
        # 速度に時間を掛けて，移動距離を計算 まず向きを計算し，その後移動距離を求める
        self.odom_mouse_angle_rad += odom_mouse_radvel * self.delta_t #マウスの角速度はパルスの左右差で求められる。現在の角度は角速度に時間をかけた角度変化量を合計していくと求まる。
        self.odom_xpos += odom_mouse_vel * math.cos(self.odom_mouse_angle_rad) * self.delta_t #向いている角度と、進んでいる速度の情報から、ロボットの現在のx座標を更新する。
        self.odom_ypos += odom_mouse_vel * math.sin(self.odom_mouse_angle_rad) * self.delta_t #y座標版
        god_mouse_pos = self.data.xpos[body_id] #シミュレーション実行前のマウスの位置を取得
        god_mouse_xpos = god_mouse_pos[0] #マウスのx座標
        god_mouse_ypos = god_mouse_pos[1] #マウスのy座標
        ####オドメトリ処理終了 つまり推定角度、xｙ座標は、それぞれ　self.odom_mouse_angle_rad, self.odom_xpos, self.odom_yposに格納されている。######################
        self.stepcount_for_resetgraph += 1
        self.mouse_position = (self.data.xpos[self.mouse_body_id][1]) #マウスの位置を取得してインスタンス変数の座標値を更新
        hit1 = self.data.sensor("HB1").data[0] #フォースセンサの値取得
        hit2 = self.data.sensor("HB2").data[0]
        terminated = False
        if hit1 > 0 or hit2 > 0: #センサがぶつかったことを検出したら倒れたということなので、
            died_reward = -1
            terminated = True #ぶつかったら終了
        velocity = self.data.sensor('Veloci').data
        # print(velocity)
        if velocity[0] >= 8 or velocity[1] >= 8 or velocity[2] >= 8:
            truncated = True
            print("シミュレーションの破綻によるぶっ飛び発生のため強制終了します")

        self.force_sensor_deque.append(hit2)
        # 平均が100以上の場合の処理
        if len(self.force_sensor_deque) > 14:  # デキューが少なすぎないことを確認
            avg = sum(self.force_sensor_deque) / len(self.force_sensor_deque)
            if avg >= 10:
                truncated = True
                print("平均が10以上です。めり込んでると思われるので強制終了します")

        self.hit_deque.append(hit2) #前方の、接触センサのdeque
        observation = self._get_obs(self.angsum_obs, self.action)

        forward_reward = self.data.sensor("Veloci").data[0]
        if sum(value != 0 for value in self.hit_deque) >= 12:
            print("0以外の値が12個以上あるので失敗！")
            died_reward = -3
            terminated = True
        if forward_reward > 1:
            forward_reward = 1 #後ろに進む場合は、罰を3倍にする
        if forward_reward < 0:
            forward_reward = forward_reward * 0.25 #後ろに進む場合は、罰を3倍にする
            
        yaw_panelty = -0.0002*abs(self.data.sensor('Gyro').data[2])
        self.zahyou_deque.append((god_mouse_xpos, god_mouse_ypos))
        # キューの中身をnumpy配列に変換
        np_zahyou_deque = np.array(self.zahyou_deque)
        # キュー内の分散を計算 (点が1つしかない場合、分散は0)
        if len(np_zahyou_deque) > 1:
            variance = np.var(np_zahyou_deque, axis=0)  # 各次元の分散
            pos_move_var_reward = 2 * np.sum(variance)  # x, y の分散の合計
            self.valiance.append(np.sum(variance))
        else:
            pos_move_var_reward = 0.0  # データが1つしかない場合は報酬ゼロ
            self.valiance.append(np.nan)

        reward = died_reward + (0.5*forward_reward) + yaw_panelty + pos_move_var_reward
        self.epi_reward += reward
        reward = float(reward)
        info = {}
        
        
            
        if self.render_mode == "human":
            self.render()
        if terminated or truncated == True:
            self.epi_reward_graph.append(self.epi_reward) #エピソードの報酬合計を追加
            self.epi_steplength_graph.append(self.stepcount) #エピソードにかかったステップ数を記録
            self.reward_average_graph.append(self.epi_reward / self.stepcount) #エピソードの報酬合計をステップ数で割って、ステップごとの報酬を記録
            self.valiance_average_graph.append(np.nanmean(self.valiance)) #そのエピソードのステップごとの分散の平均値を記録（つまり常に移動し続けていたかどうかの指標）
            self.valiance = []
            self.stepcount = 0
        return observation, reward, terminated, truncated, info
    
    def _get_obs(self, angsum_obs, action): #状態空間を取得 学習のループで呼ばれる用
        # print("obs")
        angsum_obs = np.array([angsum_obs])
        # position = self.data.qpos[0:7].flat.copy() #神の視点の位置情報取得
        qpos = self.data.qpos.flat.copy() #神の視点の位置情報取得
        qvel = self.data.qvel.flat.copy() #神の視点の位置情報取得
        body_id = mujoco.mj_name2id(self.model, mujoco.mjtObj.mjOBJ_BODY, "mouse_body")  # ボディ名からIDを取得
        yawjoint_id = mujoco.mj_name2id(self.model, mujoco.mjtObj.mjOBJ_JOINT, "yaw")  # ボディ名からIDを取得

        self.veloci_sensor_deque.append(self.data.sensor("Veloci").data[0])
        x_veloci = np.array(self.veloci_sensor_deque)
        self.acc_x_sensor_deque.append(self.data.sensor("Accel").data[0])
        acc_x = np.array(self.acc_x_sensor_deque)
        self.gyro_z_sensor_deque.append(self.data.sensor("Gyro").data[2])
        gyro_z = np.array(self.gyro_z_sensor_deque)
        # hit1 = np.array([self.data.sensor("HB1").data[0]])
        # print("action", action)
        act_Left = np.array(self.act_Left_deque)
        act_Right = np.array(self.act_Right_deque)
        self.LS_sensor_deque.append(self.data.sensor("LS").data[0])
        self.LF_sensor_deque.append(self.data.sensor("LF").data[0])
        self.RS_sensor_deque.append(self.data.sensor("RS").data[0])
        self.RF_sensor_deque.append(self.data.sensor("RF").data[0])
        sensorLS = np.array(self.LS_sensor_deque)
        sensorLF = np.array(self.LF_sensor_deque)
        sensorRS = np.array(self.RS_sensor_deque)
        sensorRF = np.array(self.RF_sensor_deque)
        return np.concatenate((sensorLS, sensorLF, sensorRS, sensorRF, x_veloci, act_Left, act_Right, acc_x, gyro_z))# newest_x, newest_y, quarter_3_x, quarter_3_y, middle_x, middle_y, quarter_1_x, quarter_1_y, oldest_x, oldest_y))   
     
    def reset_model(self):
        # print("reset!!!!!!!!!!")
        qpos = self.init_qpos.copy() #初期位置を取得
        # print("qpos", qpos)
        qvel = self.init_qvel.copy() #初期速度を取得
        self.valiance = []
        self.stepcount = 0
        goal_geom_id = mujoco.mj_name2id(self.model, mujoco.mjtObj.mjOBJ_GEOM, "goal_mark")
        body_id = mujoco.mj_name2id(self.model, mujoco.mjtObj.mjOBJ_BODY, "mouse_body")  # ボディ名からIDを取得
        
        self.acc_xsensor_history = [0, 0, 0]
        self.acc_ysensor_history = [0, 0, 0]
        self.gyro_zsensor_history = [0, 0, 0]
        self.LSsensor_history = [0, 0, 0] 
        self.LFsensor_history = [0, 0, 0]
        self.RFsensor_history = [0, 0, 0]
        self.RSsensor_history = [0, 0, 0] #各センサの履歴を初期化

        self.zahyou_deque.clear()
        self.force_sensor_deque.clear()
        self.hit_deque.clear()
        # 各観測用deque
        self.LS_sensor_deque = deque([0, 0, 0, ], maxlen=3)
        self.LF_sensor_deque = deque([0, 0, 0, ], maxlen=3)
        self.RF_sensor_deque = deque([0, 0, 0, ], maxlen=3)
        self.RS_sensor_deque = deque([0, 0, 0, ], maxlen=3)
        self.act_Left_deque = deque([0], maxlen=1)
        self.act_Right_deque = deque([0], maxlen=1)
        self.veloci_sensor_deque = deque([0], maxlen=1)
        self.acc_x_sensor_deque = deque([0], maxlen=1)
        self.acc_y_sensor_deque = deque([0, 0, 0, 0, 0], maxlen=5)
        self.gyro_z_sensor_deque = deque([0, ], maxlen=1)

        qpos[0] = (0.18*np.random.randint(0, 16)) #これxであってるかな？ ちゃんとxだった（あってた）
        qpos[1] = (0.18*np.random.randint(0, 16))
        qpos[3] = np.random.uniform(0, 3.14)
        if self.mujoco_renderer.viewer is None: #マーカーを使いたい場合この２行が必要！！！
            self.render()
        self.set_state(qpos, qvel) #qposとqvelには，すべての位置と速度の値がならんでいるためそれをセットする#################
        self.odom_mouse_angle_rad = math.pi/2 + qpos[3] #初期角度に、ランダムにずらした値を加える
        self.odom_xpos = self.data.xpos[self.mouse_body_id][0]
        self.odom_ypos = 0.18 #self.data.xpos[self.mouse_body_id][1]（つまりy座標）でも良いけど、まあ固定だし数値にしておく
        self.angsum_obs = 0
        self.epi_reward = 0 #収益を数え終わったから０にリセットして次のエピソードの報酬を数えていくため

        observation = self._get_obs(self.angsum_obs, self.action)

        return observation

    def _get_reset_info(self):
        return {"works": True}

#＃＃＃＃＃＃オプション。特に使わない＃＃＃＃＃＃＃＃＃＃＃＃
CHECK_ENV_IGNORE_WARNINGS = [
    f"\x1b[33mWARN: {message}\x1b[0m"
    for message in [
        "A Box observation space minimum value is -infinity. This is probably too low.",
        "A Box observation space maximum value is infinity. This is probably too high.",
        "For Box action spaces, we recommend using a symmetric and normalized space (range=[-1, 1] or [0, 1]). See https://stable-baselines3.readthedocs.io/en/master/guide/rl_tips.html for more information.",
    ]
]


@pytest.mark.parametrize("frame_skip", [1, 2, 3, 4, 5])
def test_frame_skip(frame_skip):
    """verify that custom envs work with different `frame_skip` values"""
    env = Mouse1massEnv(frame_skip=frame_skip) #フレームスキップ設定の上書き？

    # Test if env adheres to Gym API
    with warnings.catch_warnings(record=True) as w:
        check_env(env.unwrapped, skip_render_check=True)
        env.close()
    for warning in w:
        if warning.message.args[0] not in CHECK_ENV_IGNORE_WARNINGS:
            raise Error(f"Unexpected warning: {warning.message}")


def test_xml_file():
    """Verify that the loading of a custom XML file works \\
      訳：XMLファイルの読み込みがうまくいくことを確認する"""
    relative_path = "./tests/envs/mujoco/assets/walker2d_v5_uneven_feet.xml"
    env = Mouse1massEnv(xml_file=relative_path).unwrapped
    assert isinstance(env, MujocoEnv)
    assert env.data.qpos.size == 9

    full_path = os.getcwd() + "/tests/envs/mujoco/assets/walker2d_v5_uneven_feet.xml"
    env = Mouse1massEnv(xml_file=full_path).unwrapped
    assert isinstance(env, MujocoEnv)
    assert env.data.qpos.size == 9


def test_reset_info():
    """Verify that the environment returns info at `reset()`"""
    env = Mouse1massEnv()

    _, info = env.reset()
    assert info["works"] is True
#＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃オプションここまで＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃

#＃＃＃＃＃＃＃＃＃＃用意した環境クラスを使用し、学習済みモデルをロード＃＃＃＃＃＃＃＃＃＃＃＃

import gymnasium as gym
gym.envs.registration.register(id='Mouse1massEnv-v0',max_episode_steps=3000,entry_point=Mouse1massEnv) 
#環境登録 idは環境の名前付け。 max_episode_steps=600, は1エピソードの最大ステップ数 entry_point=MouseRightEnv は環境クラスの指定

from stable_baselines3 import SAC, PPO
from tqdm import trange
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
from gymnasium.wrappers import RecordVideo
import datetime

from stable_baselines3.common.callbacks import BaseCallback

start_time = datetime.datetime.now().replace(microsecond=0) #学習にかかった時間を計測するために使用　小数の時間は切り捨て(microsecond=0)

env = gym.make("Mouse1massEnv-v0", render_mode="human") #環境クラスの作成。レンダーモードを選択 


model = env.model
data = env.data

odom_right_graph = []
mouse_correct_vel_graph = []

gear = 9.0e-3
wheel_r = 0.0135
tread = 0.072
mx = 0.0
my = 0.0
psi = 0.0
past_odom_right = 0.0
past_odom_left = 0.0
st_Kp = 8.4
rad_Kp = 0.15
rad_vel_Kp = 2

rad_Kd = 0.00001

ki6 = 0.02
kd = 0.09
I_err_sum = 0
d_grad = 0
prev_err_rad = 0
target_vel = 0
target_rad = math.pi/2
target_rad_vel = 0
mjstep_count = 0

right_pulse_num = 0
right_pulse_num_graph = []
left_pulse_num = 0
left_pulse_num_graph = []

right_rotation_sum = 0
left_rotation_sum = 0

mouse_vel = 0
mouse_rad_vel = 0
mouse_xpos = 0
mouse_xpos_graph = []
mouse_ypos = 0
mouse_ypos_graph = []
target_vel_graph = []
err_vel_graph = []
mouse_vel_graph = []

viz_now = 0.0
viz_past = 0.0
now_time = 0
past_time = 0
turn_flag = 0
x_vel = 0
y_vel = 0
z_vel = 0
err = 0
delta_t = 0.001 ##########################制御周期を変える場合はここも変える！！！！！！
mouse_now_rad_xy = math.pi/2 #y軸方向を向いている
I_err_rad_vel_sum = 0
mouse_angle_rad = (math.pi/2)
right_mot = 0
left_mot = 0 #最初は0

#デバッグ用グラフのためのリスト
LRF_lf_graph = []
LRF_ls_graph = []
LRF_rs_graph = []
LRF_rf_graph = []
lf_sensor_graph = []
rf_sensor_graph = []
ls_sensor_graph = []
rs_sensor_graph = []
t_graph = []
kakusokudo_graph = []
rlkakusokudo_graph = []
kakudo_graph = []
final_kakudo_graph = []
rlfinal_kakudo_graph = []
rlkakudo_graph = []
right_mot_graph = []
left_mot_graph = []
x_vel_graph = []
velocimeter_graph = []
err_graph = []
err_rad_graph = []
now_kakudo_graph = []
now_kakudo_god_graph = []
kakudo_err_graph = []
target_kakudo_graph = []
gyro_graph = []
yzahyou = []
timevals = []
timevals5 = []
timevals6 = []
mouse_angle_rad_graph = []
mouse_correct_rad_vel_graph = []
mouse_rad_vel_graph = []
err_rad_vel_graph = []
target_rad_vel_graph = []
```

```python
average = []
for kk in range(10):
    #環境の定義

    __credits__ = ["Kallinteris-Andreas"]

    import os
    import warnings
    import time
    import matplotlib.pyplot as plt
    import mujoco
    import mujoco.viewer
    import glfw
    # import keyboard
    import math
    import numpy as np
    import pytest
    import random
    from gymnasium import utils
    from gymnasium.envs.mujoco import MujocoEnv
    from gymnasium.error import Error
    from gymnasium.spaces import Box
    from gymnasium import spaces
    from gymnasium.utils.env_checker import check_env
    from scipy.spatial.transform import Rotation as R
    from collections import deque




    class Mouse1massEnv(MujocoEnv, utils.EzPickle):
        """
        Gymansium.MujocoEnv`環境APIを使った，倒立振子の強化学習環境\n
        倒れないで立ち続けるタスク．\n
        mujoco envを継承しています．
        """

        metadata = {
            "render_modes": [
                "human",
                "rgb_array",
                "depth_array",
            ],
        }

        def __init__(self, xml_file="my_xmls/ex_course.xml", frame_skip=2, **kwargs):
        # def __init__(self, xml_file="my_xmls/16x16maze_model_NO_merikomi_masatu.xml", frame_skip=2, **kwargs):
            utils.EzPickle.__init__(self, xml_file, frame_skip, **kwargs)

            MujocoEnv.__init__(
                self,
                xml_file,
                frame_skip=frame_skip,
                observation_space=None,  # needs to be defined after
                default_camera_config={},
                camera_name="camera_name",
                **kwargs,
            )
            

            self.metadata = { #またメタデータがある。インスタンス変数？
                "render_modes": [
                    "human",
                    "rgb_array",
                    "depth_array",
                ],
                "render_fps": int(np.round(1.0 / self.dt)),
            }

            obs_size = 17 #self.data.qpos.size + self.data.qvel.size #たとえば，観測空間に位置と速度を入れたいのであれば，その数をここに入れる。次元数なので数え方は０始まりではない。
            # self.wall_hit = 0
            #マイクロマウスのパラメータ用変数の用意
            self.reward_graph = []
            self.deg_difference_graph = []
            self.x_velocity_reward_graph = []
            self.odom_vel_graph = []
            self.epi_reward_graph = []
            self.x_veloci_reward_graph = []

            self.kakudo_graph = []
            self.god_kakudo_graph = []
            self.god_x_veloci_graph = []
            self.god_y_veloci_graph = []
            self.god_z_veloci_graph = []
            self.x_veloci_graph = []
            self.y_veloci_graph = []
            self.z_veloci_graph = []
            self.god_x_pos_graph = []
            self.god_y_pos_graph = []
            self.god_z_pos_graph = []
            self.x_pos_graph = []
            self.y_pos_graph = []
            self.z_pos_graph = []
            self.gyro_x_graph = []
            self.gyro_y_graph = []
            self.gyro_z_graph = []
            self.acc_x_graph = []
            self.acc_y_graph = []
            self.acc_z_graph = []
            self.mouse_xpos_graph = []
            self.mouse_ypos_graph = []
            self.lastxpos_graph = []
            self.lastypos_graph = []
            self.odom_xpos_graph = []
            self.odom_ypos_graph = []
            self.god_xpos_graph = []
            self.god_ypos_graph = []
            self.odom_mouse_angle_rad = math.pi/2
            self.odom_xpos = 0.27
            self.odom_ypos = 0.18
            self.sec_0_flag = False
            self.sec_1_flag = False
            self.sec_2_flag = False
            self.sec_3_flag = False
            self.sec_4_flag = False
            self.sec_5_flag = False
            self.sec_6_flag = False
            self.sec_7_flag = False
            self.sec_8_flag = False
            self.sec_9_flag = False
            self.sec_10_flag = False
            self.sec_11_flag = False
            self.sec_12_flag = False
            self.sec_13_flag = False
            self.goal_random = 8 #適当に、0と1以外の値にした
            self.wall_mode = 0
            self.mouse_angle_rad = math.pi/2
            self.leftwall = 1
            self.rightwall = 1
            self.forwardwall = 1
            self.backwall = 1
            self.mouse_xpos = 0
            self.mouse_ypos = 0
            self.mouse_vel = 0
            self.now_time = 0
            self.prev_time = 0
            self.delta_t = 0.01
            self.x_veloci = 0
            self.y_veloci = 0
            self.z_veloci = 0
            self.x_pos = 0
            self.y_pos = 0
            self.z_pos = 0
            self.gear = 9.0e-3
            self.wheel_r = 0.0135
            self.tread = 0.072
            self.right_rotation_sum = 0
            self.left_rotation_sum = 0
            self.encoder_pulses_per_turn = 4096
            self.odom_x_graph = []
            self.odom_y_graph = []
            self.stepcount = 0
            self.stepcount_for_resetgraph = 0
            self.angsum_obs = 0
            self.epi_reward = 0
            self.reward_average_graph = []
            self.epi_steplength_graph = []
            self.valiance = []
            self.valiance_average_graph = []
            self.obsLS_graph = []
            self.obsLF_graph = []
            self.obsRS_graph = []
            self.obsRF_graph = []
            self.range_reward_x_graph = []
            self.range_reward_y_graph = []
            self.xpos_graph = []
            self.ypos_graph = []
            self.firstxyzahyou = self.init_qpos[0:3]
            self.firstquat = self.init_qpos[3:7]
            self.first_ang = (0, 0, 90)
            self.first_ang_graph = []
            self.goal_pos_number = 2
            self.geom_id1 = mujoco.mj_name2id(self.model, mujoco.mjtObj.mjOBJ_GEOM, "right_wheel")  # ジョイント名からIDを取得
            self.geom_id2 = mujoco.mj_name2id(self.model, mujoco.mjtObj.mjOBJ_GEOM, "left_wheel")  # ジョイント名からIDを取得

            self.mouse_body_id = mujoco.mj_name2id(self.model, mujoco.mjtObj.mjOBJ_BODY, "mouse_body")  # ボディ名からIDを取得
            self.mouse_position = 0

            self.hit_graph = [] #壁にぶつかったかどうかを0,1で記録するためのリスト
            self.total_distance = 0 #マウスが進んだ合計距離を記録するための変数
            self.blast_flag = 0
            self.god_z_deque = deque([0] * 200, maxlen=200)


            # 過去のセンサ値を保存するリスト
            self.acc_xsensor_history = [0, 0, 0]
            self.acc_ysensor_history = [0, 0, 0]
            self.gyro_zsensor_history = [0, 0, 0]
            self.LSsensor_history = [0, 0, 0]
            self.LFsensor_history = [0, 0, 0]
            self.RFsensor_history = [0, 0, 0]
            self.RSsensor_history = [0, 0, 0]

            # 座標分散報酬計算用キューの最大サイズを5に設定
            self.zahyou_deque = deque(maxlen=50)
            # 足跡観測用キューの最大サイズを5に設定
            # self.asiato_deque = deque(maxlen=200)
            # 壁めり込んだことを検出する用のキュー
            self.force_sensor_deque = deque(maxlen=15)
            # 壁ぶつかった後はしばらくは後退したら報酬を与えるための、速度のログdeque
            self.hit_deque = deque(maxlen=25)

            # 各観測用deque
            self.LS_sensor_deque = deque([0, 0, 0, ], maxlen=3)
            self.LF_sensor_deque = deque([0, 0, 0, ], maxlen=3)
            self.RF_sensor_deque = deque([0, 0, 0, ], maxlen=3)
            self.RS_sensor_deque = deque([0, 0, 0, ], maxlen=3)
            self.act_Left_deque = deque([0], maxlen=1)
            self.act_Right_deque = deque([0], maxlen=1)
            self.veloci_sensor_deque = deque([0], maxlen=1)
            self.acc_x_sensor_deque = deque([0], maxlen=1)
            self.acc_y_sensor_deque = deque([0, 0, 0, 0, 0], maxlen=5)
            self.gyro_z_sensor_deque = deque([0, ], maxlen=1)
                    
            self.action = [0, 0]



            self.observation_space = Box( #観測空間を定義．gymnasium.sapcesのBoxを使用することで簡単に定義できる．Boxは連続値を扱う．
                low=-np.inf, high=np.inf, shape=(obs_size,), dtype=np.float64 #float64型の，-無限から無限までの値を取る，11次元の観測空間を用意
            )

        def get_now_wheel_ang(self, model, data): #今のタイヤの回転角度と速度（角度と角速度）を取得．
            """
                現在のタイヤの回転角度と速度（と角速度）を取得する．\\
                data.actuator().length,velocityで取得できる値はactuatorのgear倍されているので，それで割ることによって打ち消す，
                Returns:
                    odm_right = 右タイヤの回転角度[rad]（基準からの総量） \\
                    odm_left = 左タイヤの回転角度[rad]（基準からの総量） \\
                    (vel_left) = 左タイヤの回転角速度 (単位時間あたり) \\
                """
            right_wheel_ang = data.actuator('right').length[0]/self.gear # 得た値をgearで割ったもの＝ギアの付いた先の、タイヤ自体の回転した角度(rad)
            left_wheel_ang = data.actuator('left').length[0]/self.gear # タイヤの角度
            vel_left = data.actuator('left').velocity[0]/self.gear # ＝タイヤの回転角速度
            return right_wheel_ang, left_wheel_ang # タイヤの角度を取得
        

        def get_pulse_count(self, pre_right_wheel_ang, pre_left_wheel_ang, right_wheel_ang, left_wheel_ang): #回転量をパルスにする エンコーダの役割をする関数
            """
                回転量の差から，出たパルスの数を求める． (エンコーダの役割をする関数)\\
                
                Returns:
                    right_pulse_num = 右タイヤのパルス数 \\
                    left_pulse_num = 左タイヤのパルス数
                """

            pre_right_rotation_sum = self.right_rotation_sum #１つまえの合計回転量取得
            pre_left_rotation_sum = self.left_rotation_sum
            self.right_rotation_sum += (right_wheel_ang - pre_right_wheel_ang)/(2*math.pi) *4096 #1周あたり4096段階で，細かい回転量を計測（分解能ほぼ無限のエンコーダ）（小数で、0から1で回転を表していたものを0から4096で表すようにした）
            self.left_rotation_sum += (left_wheel_ang - pre_left_wheel_ang)/(2*math.pi) *4096 #1周あたり4096段階で，細かい回転量を計測（分解能ほぼ無限のエンコーダ）
            right_pulse_num = int(self.right_rotation_sum) - int(pre_right_rotation_sum) #現在の回転量を足した後の回転合計値ー足す前の回転合計値　これのintが１以上なら、１パルス出たことになる。パルスが出なくても回転量の合計は溜まっていく
            left_pulse_num = int(self.left_rotation_sum) - int(pre_left_rotation_sum)
            # print("回転量をパルス段階に変換　この値の差が，実際に出たパルスとなる：", right_pulse_num)
            return right_pulse_num, left_pulse_num #出たパルスの数

        def calc_odometry_at_pulse(self, right_pulse_num, left_pulse_num): #今のタイヤの回転角度と速度（角度と角速度）を求める．そのご、タイヤの回転による移動距離と移動速度を求め、車体の速度と角速度を求める。
            """
                パルス数を元に、ロボットの速度と回転角速度を求める．（オドメトリによる推定を行う）\\
                Returns:
                    mouse_vel = ロボットの推定速度 \\
                    left_pulse_num = ロボットの推定回転速度
                """
            right_wheel_move = 2*math.pi * self.wheel_r * (right_pulse_num/self.encoder_pulses_per_turn) #２πｒ×回転量（発生パルス数÷1周のパルス数）で、タイヤの移動距離を求める．
            left_wheel_move = 2*math.pi * self.wheel_r * (left_pulse_num/self.encoder_pulses_per_turn) #２πｒ×回転数で移動距離を求める．
            right_wheel_vel = right_wheel_move/self.delta_t #移動距離を時間で割れば、単位時間あたりの移動距離、つまり速度が求められる
            left_wheel_vel = left_wheel_move/self.delta_t #移動距離から，速度を求める
            robot_vel = (right_wheel_vel + left_wheel_vel)/2 #左右のタイヤの速度の平均が、車体全体の速度
            robot_radvel = (right_wheel_vel - left_wheel_vel)/self.tread #これがマウスの旋回角速度（rad/s）左右のタイヤの速度の差をtread幅の長さで割ったものが、角速度になるため。
            return robot_vel, robot_radvel #これで，エンコーダによる各タイヤの推定速度が求められた
        
        def friction_randam_change(self):
            random_number_right = np.random.rand()
            random_number_left = np.random.rand()
            # 右タイヤの摩擦係数をランダムに変更
            if random_number_right < 0.05: #5%の確率で摩擦係数を0にする
                self.model.geom_friction[self.geom_id1][0] = 0
            else:
                self.model.geom_friction[self.geom_id1][0] = 1

            # 左タイヤの摩擦係数をランダムに変更
            if random_number_left < 0.05:
                self.model.geom_friction[self.geom_id2][0] = 0
            else:
                self.model.geom_friction[self.geom_id2][0] = 1

        def step(self, action):
            """
            シミュレーションを1step進める。 \\
            Returns: \\
                observation, 観測空間。その行動後のエージェントが見えている状態を返す \\
                reward, 報酬の合計値 \\
                terminated = False, タスクの早期達成または強制終了条件にひっかかったわけではないが、意味のない行動をし続ける場合に陥ったときなどにエピソードを終了するフラグ。本環境ではそういう場面はないため常にFalseとなる。 \\
                truncated = タスクの早期達成または強制終了が起きたときにそれを示すもの。壁にぶつかったら強制終了が起きる。ぶつかったとき以外はFalse。 \\
                info = （現状未確認。何かの情報が入っている？）\\
            """
            #変数定義
            # print("step")
            terminated = False
            truncated = False
            goal_reward = 0
            died_reward = 0
            survive_reward = 0.001 # より速くゴールに到達するために、ロスタイム報酬
            body_id = mujoco.mj_name2id(self.model, mujoco.mjtObj.mjOBJ_BODY, "mouse_body")  # ボディ名からIDを取得
            self.act_Left_deque.append(action[0])
            self.act_Right_deque.append(action[1])
            # self.action = action
            
            ####オドメトリ処理開始##################################################################################
            self.prev_time = self.data.time #シミュレーション進める前の時間を取得
            pre_right_wheel_ang, pre_left_wheel_ang = self.get_now_wheel_ang(self.model, self.data) #今のタイヤの回転角度と速度（角度と角速度）を取得．
            self.friction_randam_change() #摩擦係数をランダムに変更
            # シミュレーション前の位置取得
            before_mouse_pos = self.data.xpos[body_id]
            before_x, before_y = before_mouse_pos[0], before_mouse_pos[1]  # x, y 座標を取得
            self.do_simulation(action, self.frame_skip)####シミュレーション実行###########################################
            # シミュレーション後の位置取得
            after_mouse_pos = self.data.xpos[body_id]
            after_x, after_y = after_mouse_pos[0], after_mouse_pos[1]  # x, y 座標を取得

            # ユークリッド距離を計算（2D）
            distance = math.sqrt((after_x - before_x) ** 2 + (after_y - before_y) ** 2)
            self.total_distance += distance
            self.stepcount += 1

            self.now_time = self.data.time #1step進んだ後のシミュレーションの時間を取得
            self.delta_t = self.now_time - self.prev_time #シミュレーション1ステップにかかった時間を取得
            self.prev_time = self.now_time #シミュレーション前の時間の変数を更新
            right_wheel_ang, left_wheel_ang = self.get_now_wheel_ang(self.model, self.data)# エンコーダをもう一度読み取る　（行動後の値を取得）
            
            right_pulse_num, left_pulse_num = self.get_pulse_count(pre_right_wheel_ang, pre_left_wheel_ang, right_wheel_ang, left_wheel_ang) #シミュレーション前後のエンコーダ値から，この一瞬に出たパルス数を算出し，
            odom_mouse_vel, odom_mouse_radvel = self.calc_odometry_at_pulse(right_pulse_num, left_pulse_num) # パルス数から，移動速度と角速度の推定値を求める。　これで，エンコーダによるマウスの速度・角速度が推定できた
            # 速度に時間を掛けて，移動距離を計算 まず向きを計算し，その後移動距離を求める
            self.odom_mouse_angle_rad += odom_mouse_radvel * self.delta_t #マウスの角速度はパルスの左右差で求められる。現在の角度は角速度に時間をかけた角度変化量を合計していくと求まる。
            self.odom_xpos += odom_mouse_vel * math.cos(self.odom_mouse_angle_rad) * self.delta_t #向いている角度と、進んでいる速度の情報から、ロボットの現在のx座標を更新する。
            self.odom_ypos += odom_mouse_vel * math.sin(self.odom_mouse_angle_rad) * self.delta_t #y座標版
            # self.odom_xpos_graph.append(self.odom_xpos)
            # self.odom_ypos_graph.append(self.odom_ypos)
            god_mouse_pos = self.data.xpos[body_id] #シミュレーション実行前のマウスの位置を取得
            god_mouse_xpos = god_mouse_pos[0] #マウスのx座標
            god_mouse_ypos = god_mouse_pos[1] #マウスのy座標
            # self.god_xpos_graph.append(god_mouse_xpos)
            # self.god_ypos_graph.append(god_mouse_ypos)
            ####オドメトリ処理終了 つまり推定角度、xｙ座標は、それぞれ　self.odom_mouse_angle_rad, self.odom_xpos, self.odom_yposに格納されている。######################
            self.stepcount_for_resetgraph += 1
            self.mouse_position = (self.data.xpos[self.mouse_body_id][1]) #マウスの位置を取得してインスタンス変数の座標値を更新
            hit1 = self.data.sensor("HB1").data[0] #フォースセンサの値取得
            hit2 = self.data.sensor("HB2").data[0]
            terminated = False
            if hit1 > 0 or hit2 > 0: #センサがぶつかったことを検出したら倒れたということなので、
                died_reward = -1
                self.hit_graph.append(1)
                terminated = False #ぶつかっても、死ななくした
            else:
                self.hit_graph.append(0)
            velocity = self.data.sensor('Veloci').data
            # print(velocity)
            self.blast_flag = 0
            # if velocity[0] >= 2 or velocity[1] >= 2 or velocity[2] >= 2:
            #     self.blast_flag = 1
            #     truncated = True
            #     print("シミュレーションの破綻によるぶっ飛び発生のため強制終了します")
            
            # **新しい値を少数第2桁に丸めて deque に保存**
            self.god_z_deque.append(round(god_mouse_pos[2], 5))

            # **すべての値が同じなら条件に入る**
            if len(set(self.god_z_deque)) == 1:
                # print("すべての値が同じ！")
                self.blast_flag = 1

            self.force_sensor_deque.append(hit2)
            # 平均が100以上の場合の処理
            # if len(self.force_sensor_deque) > 14:  # デキューが少なすぎないことを確認
            #     avg = sum(self.force_sensor_deque) / len(self.force_sensor_deque)
            #     if avg >= 10:
                    # truncated = True
                    # print("平均が10以上です。めり込んでると思われるので強制終了します")
            
            # mouse_pos = self.data.xpos[body_id]
            # # print("mouse_pos", mouse_pos)
            # self.mouse_xpos = mouse_pos[0]
            # self.mouse_ypos = mouse_pos[1]
            # # self.odom_x_graph.append(self.mouse_xpos)
            # # self.odom_x_graph.append(self.mouse_ypos)
            # yawjoint_id = mujoco.mj_name2id(self.model, mujoco.mjtObj.mjOBJ_JOINT, "yaw")  # ボディ名からIDを取得
            # mouse_yaw = self.data.qpos[yawjoint_id] #マウスの向いている角度を取得

            # print("mouse_xpos", self.mouse_xpos)
            # print("mouse_ypos", self.mouse_ypos)
            # if self.mouse_xpos < 0.01 or self.mouse_xpos > 0.38 or self.mouse_ypos < 0.16 or self.mouse_ypos > 0.38: #範囲外に出たら (ゴールよりもすこし大きめに範囲を取ってる)
            #     died_reward = -1
            #     terminated = True

            # geom_id = mujoco.mj_name2id(self.model, mujoco.mjtObj.mjOBJ_GEOM, "goal_mark")
            # goal_xpos = self.data.geom_xpos[geom_id][0]
            # goal_ypos = self.data.geom_xpos[geom_id][1]
            # goal_range = np.sqrt((mouse_xpos - goal_xpos)**2 + (mouse_ypos - goal_ypos)**2) #ロボットとゴールの距離を計算
            # goal_range_reward = -goal_range #ゴールとの距離が近いほど報酬が高い

            # mokuhyou_ang = np.array([1, 0, 0, 0]) ## 目標角度設定　　クオータニオンで表す。　右向き：[1, 0, 0, 0] 左向き：[6.123233995736766e-17 , 0, 0, 1] #正面：[0.70710678 0.         0.         0.70710678]
            # x_veloci_reward = 0 * 1/1000 * self.data.sensor("Veloci").data[0]**2 #速度の二乗が負の報酬
            # print(mouse_pos)
            # self.x_veloci_reward_graph.append(x_veloci_reward)

            self.hit_deque.append(hit2) #前方の、接触センサのdeque
            observation = self._get_obs(self.angsum_obs, self.action)

            # vel_left = data.actuator('left').velocity[0]/self.gear # ＝タイヤの回転角速度
            # vel_right = data.actuator('right').velocity[0]/self.gear # ＝タイヤの回転角速度
            # forward_reward = (vel_left + vel_right) / 2 #タイヤの回転角速度が報酬、負の場合は負
            forward_reward = self.data.sensor("Veloci").data[0]
            # print("forward_reward", forward_reward)
            # 0以外の値を数えて12以上かをチェック
            if sum(value != 0 for value in self.hit_deque) >= 12:
                # print("0以外の値が12個以上あるので失敗！")
                died_reward = -3
                # terminated = True
            # print("veloci:", forward_reward)
            if forward_reward > 1:
                forward_reward = 1 #後ろに進む場合は、罰を3倍にする
            if forward_reward < 0:
                forward_reward = forward_reward * 0.25 #後ろに進む場合は、罰を3倍にする
                # # 0以上の値が存在するかをチェック
                # if any(value > 0 for value in self.hit_deque): #valueよりookiiい値があったら、
                #     forward_reward * (-2.5) # usironiikubeki
                
            yaw_panelty = -0.0002*abs(self.data.sensor('Gyro').data[2])
            # self.asiato_deque.append((god_mouse_xpos, god_mouse_ypos)) #まずはオドメトリではなく真の値
            #位置座標の分散を計算　　①まずはそのまま追加
            self.zahyou_deque.append((god_mouse_xpos, god_mouse_ypos))
            # キューの中身をnumpy配列に変換
            np_zahyou_deque = np.array(self.zahyou_deque)
            # キュー内の分散を計算 (点が1つしかない場合、分散は0)
            if len(np_zahyou_deque) > 1:
                variance = np.var(np_zahyou_deque, axis=0)  # 各次元の分散
                pos_move_var_reward = 2 * np.sum(variance)  # x, y の分散の合計
                self.valiance.append(np.sum(variance))
            else:
                pos_move_var_reward = 0.0  # データが1つしかない場合は報酬ゼロ
                self.valiance.append(np.nan)

            # print("分散報酬：", pos_move_var_reward)
            # survive_reward = 0.001 # より速くゴールに到達するために、ロスタイム報酬
            # マスの中央を通るほどでかい報酬
            kukaku_X = int(god_mouse_xpos / 0.18) # 自分の座標から，今いるマスを計算
            kukaku_Y = int(god_mouse_ypos / 0.18) #
            # print("kukakuX:", kukaku_X, "kukakuY:", kukaku_Y)
            center_X = (kukaku_X + 0.5) * 0.18 # そのマスの中心座標を計算
            center_Y = (kukaku_Y + 0.5) * 0.18 # その中心と自分の座標との距離で，報酬を計算
            # print("centerX:", center_X, "centerY:", center_Y)
            dx = abs(god_mouse_xpos - center_X)# その中心と自分の座標との距離で，報酬を計算
            dy = abs(god_mouse_ypos - center_Y)
            distance = np.sqrt(dx**2 + dy**2)  # ユークリッド距離
            max_distance = np.sqrt(2) * (0.18 / 2) # 最大距離（セルの対角線の半分）
            center_kukaku_reward = 1 - (distance / max_distance)# 報酬計算（中心に近いほど1に、端にいると0になる）
            center_kukaku_reward = max(center_kukaku_reward, 0)# 0以下にならないように
            # print(center_kukaku_reward)


            # wall_range = self.data.sensor("1").data[0]+self.data.sensor("2").data[0]+self.data.sensor("3").data[0]+self.data.sensor("4").data[0]+self.data.sensor("5").data[0]+self.data.sensor("6").data[0]+self.data.sensor("7").data[0]+self.data.sensor("8").data[0]+self.data.sensor("9").data[0]+self.data.sensor("10").data[0]+self.data.sensor("11").data[0]+self.data.sensor("12").data[0]+self.data.sensor("13").data[0]+self.data.sensor("14").data[0]+self.data.sensor("15").data[0]+self.data.sensor("16").data[0]+self.data.sensor("17").data[0]+self.data.sensor("18").data[0]+self.data.sensor("19").data[0]+self.data.sensor("20").data[0]+self.data.sensor("21").data[0]+self.data.sensor("22").data[0]+self.data.sensor("23").data[0]+self.data.sensor("24").data[0]
            # wall_reward = wall_range * 0.005
            # print("wall_reward", wall_reward)
            reward = died_reward + (0.5*forward_reward) + yaw_panelty + pos_move_var_reward + 0.1*center_kukaku_reward #+ wall_reward # + goal_range_reward + vectol_reward# - 0* goal_range_reward + died_reward + deg_difference_reward + 0*x_veloci_reward- 0.1* np.sum(np.square(action)) #+ 0.01*x_veloci_reward #+ x_velocity_reward# - 0.01* np.sum(np.square(action)) #- 0.001 *ang_error - 0.001*Z_angvel
            self.epi_reward += reward
            reward = float(reward)
            info = {}
            
            
                
            if self.render_mode == "human":
                # self.mujoco_renderer.viewer.add_marker( #➊MujocoRendererのインスタンスを作成。➋必要なモードでrenderメソッドを呼んで、viewerを初期化。➌renderer.viewer.add_marker(...)でマーカーを追加。
                #     pos=[0, 0, 0], size=[0.05, 0.05, 0.1], rgba=[1, 0, 0, 1], type=mujoco.mjtGeom.mjGEOM_BOX, label="goal")
                self.render()
            # print("reward", reward)
            if terminated or truncated == True:
                self.epi_reward_graph.append(self.epi_reward) #エピソードの報酬合計を追加
                self.epi_steplength_graph.append(self.stepcount) #エピソードにかかったステップ数を記録
                self.reward_average_graph.append(self.epi_reward / self.stepcount) #エピソードの報酬合計をステップ数で割って、ステップごとの報酬を記録
                self.valiance_average_graph.append(np.nanmean(self.valiance)) #そのエピソードのステップごとの分散の平均値を記録（つまり常に移動し続けていたかどうかの指標）
                self.valiance = []
                self.stepcount = 0
            return observation, reward, terminated, truncated, info
        
        def _get_obs(self, angsum_obs, action): #状態空間を取得 学習のループで呼ばれる用
            # print("obs")
            angsum_obs = np.array([angsum_obs])
            # position = self.data.qpos[0:7].flat.copy() #神の視点の位置情報取得
            qpos = self.data.qpos.flat.copy() #神の視点の位置情報取得
            qvel = self.data.qvel.flat.copy() #神の視点の位置情報取得
            body_id = mujoco.mj_name2id(self.model, mujoco.mjtObj.mjOBJ_BODY, "mouse_body")  # ボディ名からIDを取得
            yawjoint_id = mujoco.mj_name2id(self.model, mujoco.mjtObj.mjOBJ_JOINT, "yaw")  # ボディ名からIDを取得
            # mouse_pos = self.data.xpos[body_id] #マウスの位置を取得
            # add_dispersion = np.random.normal(loc=0, scale=0.1)
            # mouse_yaw = np.array([self.data.qpos[yawjoint_id] + 0*np.random.normal(loc=5, scale=0.2)]) #マウスの角度を取得
            # odom_yaw = np.array([self.odom_mouse_angle_rad])
            # # print("body_id:", body_id)
            # # print("mouse_pos", mouse_pos)
            # odom_x = np.array([self.odom_xpos])
            # # self.odom_x_graph.append(odom_x)
            # odom_y = np.array([self.odom_ypos])
            # # self.odom_y_graph.append(odom_y)
            # mouse_xpos = np.array([mouse_pos[0] + 0*np.random.normal(loc=5, scale=0.2)]) #マウスの位置を取得
            # mouse_ypos = np.array([mouse_pos[1] + 0*np.random.normal(loc=5, scale=0.2)]) #マウスの位置を取得

            # self.mouse_position = (self.data.xpos[self.mouse_body_id]) #マウスの位置を取得
            # LS_sensor = np.array([self.data.sensor("LS").data[0]]) #左前センサの値取得
            # LF_sensor = np.array([self.data.sensor("LF").data[0]]) #左後ろセンサの値取得
            # RS_sensor = np.array([self.data.sensor("RS").data[0]]) #右前センサの値取得
            # RF_sensor = np.array([self.data.sensor("RF").data[0]]) #右後ろセンサの値取得
            # gyro = np.array([self.data.sensor('Gyro').data])
            # gyro_x = np.array([gyro[0][0]])
            # gyro_y = np.array([gyro[0][1]])
            # gyro_z = np.array([gyro[0][2]])
            # self.gyro_x_graph.append(gyro_x)
            # self.gyro_y_graph.append(gyro_y)
            # self.gyro_z_graph.append(gyro_z)
            # accelerometer = np.array([self.data.sensor('Accel').data])
            # accelerometer_x = np.array([accelerometer[0][0]])
            # accelerometer_y = np.array([accelerometer[0][1]])
            # accelerometer_z = np.array([accelerometer[0][2]])
            # self.acc_x_graph.append(accelerometer_x)
            # self.acc_y_graph.append(accelerometer_y)
            # self.acc_z_graph.append(accelerometer_z)
            # wall_info = np.array([self.leftwall, self.rightwall, self.forwardwall]) #壁情報取得、つまり地図を保持しているのと同じ
            # wall_info = np.array([self.wall_mode])
            # goal_pos_info = np.array([self.goal_pos_number])
            # self.x_veloci += accelerometer[0][0] * self.delta_t           #なぜか正負が逆。。　Vk = Vk-1 + ak * Δt(オイラー法で積分して加速度から速度を求める).
            # self.y_veloci += accelerometer[0][1] * self.delta_t
            # self.z_veloci += accelerometer[0][2] * self.delta_t
            # self.x_pos += self.x_veloci * self.delta_t #Xk = Xk-1 + Vk * Δt(オイラー法で積分して速度から位置を求める)
            # self.x_veloci_graph.append(self.x_veloci)
            # self.y_veloci_graph.append(self.y_veloci)
            # self.z_veloci_graph.append(self.z_veloci)
            # 新しいデータを履歴に追加し、古いデータを削除
            # self.acc_xsensor_history.append(accelerometer[0][0])
            # self.acc_ysensor_history.append(accelerometer[0][1])
            # self.gyro_zsensor_history.append(gyro[0][2])
            # self.LSsensor_history.append(self.data.sensor("LS").data[0])
            # self.LFsensor_history.append(self.data.sensor("LF").data[0])
            # self.RFsensor_history.append(self.data.sensor("RF").data[0])
            # self.RSsensor_history.append(self.data.sensor("RS").data[0])
            
            # if len(self.acc_xsensor_history) > 3: #サイズが決まりを超えたら
            #     self.acc_xsensor_history.pop(0) #古いデータを削除
            # acc_x_0 = np.array([self.acc_xsensor_history[0]])
            # acc_x_1 = np.array([self.acc_xsensor_history[1]])
            # acc_x_2 = np.array([self.acc_xsensor_history[2]])
            # if len(self.acc_ysensor_history) > 3: #サイズが決まりを超えたら
            #     self.acc_ysensor_history.pop(0) #古いデータを削除
            # acc_y_0 = np.array([self.acc_ysensor_history[0]])
            # acc_y_1 = np.array([self.acc_ysensor_history[1]])
            # acc_y_2 = np.array([self.acc_ysensor_history[2]])
            # if len(self.gyro_zsensor_history) > 3: #サイズが決まりを超えたら
            #     self.gyro_zsensor_history.pop(0) #古いデータを削除
            # gyro_z_0 = np.array([self.gyro_zsensor_history[0]])
            # gyro_z_1 = np.array([self.gyro_zsensor_history[1]])
            # gyro_z_2 = np.array([self.gyro_zsensor_history[2]])
            # if len(self.LSsensor_history) > 3: #サイズが決まりを超えたら
            #     self.LSsensor_history.pop(0) #古いデータを削除
            # LS_0 = np.array([self.LSsensor_history[0]])
            # LS_1 = np.array([self.LSsensor_history[1]])
            # LS_2 = np.array([self.LSsensor_history[2]])
            # if len(self.LFsensor_history) > 3: #サイズが決まりを超えたら
            #     self.LFsensor_history.pop(0) #古いデータを削除
            # LF_0 = np.array([self.LFsensor_history[0]])
            # LF_1 = np.array([self.LFsensor_history[1]])
            # LF_2 = np.array([self.LFsensor_history[2]])
            # if len(self.RFsensor_history) > 3: #サイズが決まりを超えたら
            #     self.RFsensor_history.pop(0) #古いデータを削除
            # RF_0 = np.array([self.RFsensor_history[0]])
            # RF_1 = np.array([self.RFsensor_history[1]])
            # RF_2 = np.array([self.RFsensor_history[2]])
            # if len(self.RSsensor_history) > 3: #サイズが決まりを超えたら
            #     self.RSsensor_history.pop(0) #古いデータを削除
            # RS_0 = np.array([self.RSsensor_history[0]])
            # RS_1 = np.array([self.RSsensor_history[1]])
            # RS_2 = np.array([self.RSsensor_history[2]])

            self.veloci_sensor_deque.append(self.data.sensor("Veloci").data[0])
            x_veloci = np.array(self.veloci_sensor_deque)
            self.acc_x_sensor_deque.append(self.data.sensor("Accel").data[0])
            acc_x = np.array(self.acc_x_sensor_deque)
            self.gyro_z_sensor_deque.append(self.data.sensor("Gyro").data[2])
            gyro_z = np.array(self.gyro_z_sensor_deque)
            # hit1 = np.array([self.data.sensor("HB1").data[0]])
            # print("action", action)
            act_Left = np.array(self.act_Left_deque)
            act_Right = np.array(self.act_Right_deque)
            self.LS_sensor_deque.append(self.data.sensor("LS").data[0])
            self.LF_sensor_deque.append(self.data.sensor("LF").data[0])
            self.RS_sensor_deque.append(self.data.sensor("RS").data[0])
            self.RF_sensor_deque.append(self.data.sensor("RF").data[0])
            sensorLS = np.array(self.LS_sensor_deque)
            sensorLF = np.array(self.LF_sensor_deque)
            sensorRS = np.array(self.RS_sensor_deque)
            sensorRF = np.array(self.RF_sensor_deque)
            # # 足跡観測
            # # dequeをリストに変換
            # asiato_list = list(self.asiato_deque)
            # # dequeの長さ
            # n = len(asiato_list)
            # # 必要な値を新しい順に取得
            # newest = asiato_list[-1]  # 一番新しい値
            # quarter_3 = asiato_list[3 * n // 4]  # 全体の4分の3番目に新しい値
            # middle = asiato_list[n // 2]  # 2分の1番目に新しい値
            # quarter_1 = asiato_list[n // 4]  # 全体の4分の1番目に新しい値
            # oldest = asiato_list[0]  # 一番古い値

            # newest_x = np.array([newest[0]])
            # newest_y = np.array([newest[1]])
            # quarter_3_x = np.array([quarter_3[0]])
            # quarter_3_y = np.array([quarter_3[1]])
            # middle_x = np.array([middle[0]])
            # middle_y = np.array([middle[1]])
            # quarter_1_x = np.array([quarter_1[0]])
            # quarter_1_y = np.array([quarter_1[1]])
            # oldest_x = np.array([oldest[0]])
            # oldest_y = np.array([oldest[1]])
            
            # self.x_pos_graph.append(self.x_pos)
            # robot_pos = self.x_pos
            # x_pos = np.array([self.x_pos]) #obsで結合するためにnp.arrayに変換
            # x_veloci = np.array([self.x_veloci]) #obsで結合するためにnp.arrayに変換
            # return np.concatenate((qpos, qvel, LS_sensor, LF_sensor, RS_sensor, RF_sensor, gyro_x, gyro_y, gyro_z, accelerometer_x, accelerometer_y, accelerometer_z, wall_info, goal_pos_info)) #np.arrayに変換して返す
            # return np.concatenate((mouse_xpos, mouse_ypos, mouse_yaw, LS_sensor, LF_sensor, RS_sensor, RF_sensor, gyro_x, gyro_y, gyro_z, accelerometer_x, accelerometer_y, accelerometer_z, wall_info, goal_pos_info)) #np.arrayに変換して返す
            # return np.concatenate((odom_x, odom_y, odom_yaw)) #np.arrayに変換して返す
            # return np.concatenate((LS_sensor, LF_sensor, RS_sensor, RF_sensor))  
            # print((LS_0, LS_1, LS_2, LF_0, LF_1, LF_2, RS_0, RS_1, RS_2, RF_0, RF_1, RF_2, x_veloci, action_left, action_right, newest_x, newest_y, quarter_3_x, quarter_3_y, middle_x, middle_y, quarter_1_x, quarter_1_y, oldest_x, oldest_y))
            # print((sensorLS, sensorLF, sensorRS, sensorRF, x_veloci, act_Left, act_Right, acc_x))
            return np.concatenate((sensorLS, sensorLF, sensorRS, sensorRF, x_veloci, act_Left, act_Right, acc_x, gyro_z))# newest_x, newest_y, quarter_3_x, quarter_3_y, middle_x, middle_y, quarter_1_x, quarter_1_y, oldest_x, oldest_y))   
        
        def reset_model(self):
            # print("reset!!!!!!!!!!")
            qpos = self.init_qpos.copy() #初期位置を取得
            # print("qpos", qpos)
            qvel = self.init_qvel.copy() #初期速度を取得
            # self.epi_steplength_graph.append(self.stepcount) #1step no nagasa wo tasu
            # self.valiance_average_graph.append(np.nanmean(self.valiance))
            self.valiance = []
            self.stepcount = 0
            goal_geom_id = mujoco.mj_name2id(self.model, mujoco.mjtObj.mjOBJ_GEOM, "goal_mark")
            body_id = mujoco.mj_name2id(self.model, mujoco.mjtObj.mjOBJ_BODY, "mouse_body")  # ボディ名からIDを取得
            
            self.acc_xsensor_history = [0, 0, 0]
            self.acc_ysensor_history = [0, 0, 0]
            self.gyro_zsensor_history = [0, 0, 0]
            self.LSsensor_history = [0, 0, 0] 
            self.LFsensor_history = [0, 0, 0]
            self.RFsensor_history = [0, 0, 0]
            self.RSsensor_history = [0, 0, 0] #各センサの履歴を初期化

            self.zahyou_deque.clear()
            self.force_sensor_deque.clear()
            self.hit_deque.clear()
            # 各観測用deque
            self.LS_sensor_deque = deque([0, 0, 0, ], maxlen=3)
            self.LF_sensor_deque = deque([0, 0, 0, ], maxlen=3)
            self.RF_sensor_deque = deque([0, 0, 0, ], maxlen=3)
            self.RS_sensor_deque = deque([0, 0, 0, ], maxlen=3)
            self.act_Left_deque = deque([0], maxlen=1)
            self.act_Right_deque = deque([0], maxlen=1)
            self.veloci_sensor_deque = deque([0], maxlen=1)
            self.acc_x_sensor_deque = deque([0], maxlen=1)
            self.acc_y_sensor_deque = deque([0, 0, 0, 0, 0], maxlen=5)
            self.gyro_z_sensor_deque = deque([0, ], maxlen=1)

            # self.asiato_deque.clear()
            # print("reset",goal_geom_id)
            # ランダムな位置に配置する
            # self.model.geom_pos[goal_geom_id] = 0.36, 0.27, 0.1

            qpos[0] = (0.18*np.random.randint(0, 16)) #これxであってるかな？ ちゃんとxだった（あってた）
            qpos[1] = (0.18*np.random.randint(0, 16))
            qpos[3] = np.random.uniform(0, 3.14)
            # self.asiato_deque.append((qpos[0], qpos[1]))
            # mujoco.mj_forward(self.model, self.data)    # 再計算
            # viewerが初期化されているか確認し、初期化されていなければ初期化する
            if self.mujoco_renderer.viewer is None: #マーカーを使いたい場合この２行が必要！！！
                self.render()
            # print("geom_id", geom_id)
            # if self.goal_random == 0: 
            #     self.model.geom_pos[geom_id] = 0.36, 0.27, 0.1 #ゴールを右に配置
            # elif self.goal_random == 1:
            #     self.model.geom_pos[geom_id] = 0.36, 0.36, 0.1 #ゴールを前に配置
            # mujoco.mj_forward(self.model, self.data)    # 再計算
            #壁パターンの番号を取得
            # self.wall_mode = np.random.randint(1, 7)# 1～6の整数をランダムに取得

            # # ランダムな傾きを生成（約±5度）
            # random_tilt = np.random.normal(-8, 8)
            # # 回転軸（この場合、Y軸周り）
            # axis = [1, 0, 0]
            # # 回転をクォータニオンに変換
            # r = R.from_rotvec(np.radians(random_tilt) * np.array(axis))
            # random_quaternion = r.as_quat()
            # # qposの更新（最後の4要素がクォータニオン）
            # qpos[3:7] = random_quaternion
            # print(qpos)

            # x_values = [0.27, 1.07, 1.87]
            # mouse_xposition_shuffle = random.choice(x_values)
            # x_values = 0.27

            # mouse_xposition_shuffle = self.course_list[np.random]
            # mouse_yposition_shuffle = self.course_list[np.random]

            #両方3のときはやり直しする処理
            
            # qpos[0] = mouse_xposition_shuffle
            # qpos[0] = x_values
            # noise = np.random.normal(0, 0.3)  # 平均0、標準偏差0.1のノイズ
            # qpos[3] += np.random.normal(0, 0.03)#0.27 1.07
            # qpos[1] = mouse_yposition_shuffle
            self.set_state(qpos, qvel) #qposとqvelには，すべての位置と速度の値がならんでいるためそれをセットする#################
            # self.odom_mouse_angle_rad = math.pi/2
            # self.odom_xpos = 0.27
            # self.odom_ypos = 0.18
            self.odom_mouse_angle_rad = math.pi/2 + qpos[3] #初期角度に、ランダムにずらした値を加える
            self.odom_xpos = self.data.xpos[self.mouse_body_id][0]
            self.odom_ypos = 0.18 #self.data.xpos[self.mouse_body_id][1]（つまりy座標）でも良いけど、まあ固定だし数値にしておく
            self.angsum_obs = 0

            # if self.stepcount_for_resetgraph != 0:
            #     self.epi_reward_graph.append(self.epi_reward) #収益のグラフ表示用
            self.epi_reward = 0 #収益を数え終わったから０にリセットして次のエピソードの報酬を数えていくため

            observation = self._get_obs(self.angsum_obs, self.action)
            #必要に応じてオドメトリ値もリセットする必要がある
            # print(qpos)

            return observation

        def _get_reset_info(self):
            return {"works": True}
        
        # from typing import Any, Dict, Optional, Tuple, Union
        # def reset(
        #     self,
        #     *,
        #     seed: Optional[int] = None,
        #     options: Optional[dict] = None,
        # ):
        #     super().reset(seed=seed) #サブクラスにresetを書くとそちらだけが実行されるように隠蔽されてしまうので，スーパークラスのresetを実行して，前に定義した内容も実行する

        #     self._reset_simulation()
        #     print("リセットメソッドのオーバーライドができているか確認するためのprint文です．")

        #     ob = self.reset_model()
        #     info = self._get_reset_info()

        #     if self.render_mode == "human":
        #         self.render()
        #     return ob, info

    #＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃オプション。特に使わない＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃
    CHECK_ENV_IGNORE_WARNINGS = [
        f"\x1b[33mWARN: {message}\x1b[0m"
        for message in [
            "A Box observation space minimum value is -infinity. This is probably too low.",
            "A Box observation space maximum value is infinity. This is probably too high.",
            "For Box action spaces, we recommend using a symmetric and normalized space (range=[-1, 1] or [0, 1]). See https://stable-baselines3.readthedocs.io/en/master/guide/rl_tips.html for more information.",
        ]
    ]


    @pytest.mark.parametrize("frame_skip", [1, 2, 3, 4, 5])
    def test_frame_skip(frame_skip):
        """verify that custom envs work with different `frame_skip` values"""
        env = Mouse1massEnv(frame_skip=frame_skip) #フレームスキップ設定の上書き？

        # Test if env adheres to Gym API
        with warnings.catch_warnings(record=True) as w:
            check_env(env.unwrapped, skip_render_check=True)
            env.close()
        for warning in w:
            if warning.message.args[0] not in CHECK_ENV_IGNORE_WARNINGS:
                raise Error(f"Unexpected warning: {warning.message}")


    def test_xml_file():
        """Verify that the loading of a custom XML file works \\
        訳：XMLファイルの読み込みがうまくいくことを確認する"""
        relative_path = "./tests/envs/mujoco/assets/walker2d_v5_uneven_feet.xml"
        env = Mouse1massEnv(xml_file=relative_path).unwrapped
        assert isinstance(env, MujocoEnv)
        assert env.data.qpos.size == 9

        full_path = os.getcwd() + "/tests/envs/mujoco/assets/walker2d_v5_uneven_feet.xml"
        env = Mouse1massEnv(xml_file=full_path).unwrapped
        assert isinstance(env, MujocoEnv)
        assert env.data.qpos.size == 9

        # note can not test user home path (with '~') because github CI does not have a home folder　訳：ユーザーホームパス（'〜'を含む）をテストできないことに注意してください。github CIにはホームフォルダがありません


    def test_reset_info():
        """Verify that the environment returns info at `reset()`"""
        env = Mouse1massEnv()

        _, info = env.reset()
        assert info["works"] is True
    #＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃オプションここまで＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃

    #＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃用意した環境クラスを使用し、学習済みモデルをロード＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃

    import gymnasium as gym
    gym.envs.registration.register(id='Mouse1massEnv-v0',max_episode_steps=7500,entry_point=Mouse1massEnv) #環境登録 idは環境の名前付け。 max_episode_steps=600, は1エピソードの最大ステップ数 entry_point=MouseRightEnv は環境クラスの指定

    from stable_baselines3 import SAC, PPO
    from tqdm import trange
    import matplotlib.pyplot as plt
    from matplotlib.ticker import MultipleLocator
    from gymnasium.wrappers import RecordVideo
    import datetime

    from stable_baselines3.common.callbacks import BaseCallback

    start_time = datetime.datetime.now().replace(microsecond=0) #学習にかかった時間を計測するために使用　小数の時間は切り捨て(microsecond=0)

    env = gym.make("Mouse1massEnv-v0", render_mode="human") #環境クラスの作成。レンダーモードを選択 







    #Create model
    # model = mujoco.MjModel.from_xml_path('../mouse_in_maze.xml')
    # data = mujoco.MjData(model)
    model = env.model
    data = env.data

    odom_right_graph = []
    mouse_correct_vel_graph = []

    gear = 9.0e-3
    wheel_r = 0.0135
    tread = 0.072
    mx = 0.0
    my = 0.0
    psi = 0.0
    past_odom_right = 0.0
    past_odom_left = 0.0
    # kp = 100
    # st_Kp = 1.55 #ダンピング3.6e-6のときの最適なゲイン
    st_Kp = 8.4
    rad_Kp = 0.15
    rad_vel_Kp = 2

    rad_Kd = 0.00001

    ki6 = 0.02
    kd = 0.09
    I_err_sum = 0
    d_grad = 0
    prev_err_rad = 0
    target_vel = 0
    target_rad = math.pi/2
    target_rad_vel = 0
    mjstep_count = 0

    right_pulse_num = 0
    right_pulse_num_graph = []
    left_pulse_num = 0
    left_pulse_num_graph = []

    right_rotation_sum = 0
    left_rotation_sum = 0

    mouse_vel = 0
    mouse_rad_vel = 0
    mouse_xpos = 0
    mouse_xpos_graph = []
    mouse_ypos = 0
    mouse_ypos_graph = []
    target_vel_graph = []
    err_vel_graph = []
    mouse_vel_graph = []

    viz_now = 0.0
    viz_past = 0.0
    now_time = 0
    past_time = 0
    turn_flag = 0
    x_vel = 0
    y_vel = 0
    z_vel = 0
    err = 0
    delta_t = 0.001 #############################################制御周期を変える場合はここも変える！！！！！！
    mouse_now_rad_xy = math.pi/2 #y軸方向を向いている
    I_err_rad_vel_sum = 0
    mouse_angle_rad = (math.pi/2)
    right_mot = 0
    left_mot = 0 #最初は0

    #デバッグ用グラフのためのリスト
    LRF_lf_graph = []
    LRF_ls_graph = []
    LRF_rs_graph = []
    LRF_rf_graph = []
    lf_sensor_graph = []
    rf_sensor_graph = []
    ls_sensor_graph = []
    rs_sensor_graph = []
    t_graph = []
    kakusokudo_graph = []
    rlkakusokudo_graph = []
    kakudo_graph = []
    final_kakudo_graph = []
    rlfinal_kakudo_graph = []
    rlkakudo_graph = []
    right_mot_graph = []
    left_mot_graph = []
    x_vel_graph = []
    velocimeter_graph = []
    err_graph = []
    err_rad_graph = []
    now_kakudo_graph = []
    now_kakudo_god_graph = []
    kakudo_err_graph = []
    target_kakudo_graph = []
    gyro_graph = []
    yzahyou = []
    timevals = []
    timevals5 = []
    timevals6 = []
    mouse_angle_rad_graph = []
    mouse_correct_rad_vel_graph = []
    mouse_rad_vel_graph = []
    err_rad_vel_graph = []
    target_rad_vel_graph = []

    # 学習
    import gymnasium as gym
    from stable_baselines3 import SAC, PPO ,DQN
    # from sb3_contrib import RecurrentPPO
    from tqdm import trange
    import matplotlib.pyplot as plt
    from gymnasium.wrappers import RecordVideo
    import datetime
    from stable_baselines3.common.env_util import make_vec_env
    import torch.nn as nn
    from torchviz import make_dot
    import torch
    # from matplotlib.lines import Line2D

    # env = gym.make("ALE/Breakout-v5",  render_mode="human")
    # env = gym.make("CartPole-v1",  render_mode="human")
    # # env = Monitor(env, "./gym-results", force=True, video_callable=lambda episode: True)　こんな感じで，
    # model = SAC("MlpPolicy", env, verbose=1)
    # model.learn(total_timesteps=200)
    from stable_baselines3.common.callbacks import BaseCallback

    # start_time = datetime.datetime.now().replace(microsecond=0)
    # 環境を作成する関数を定義
    # def make_env():
    #     return gym.make("MyCartPoleEnv-v0", render_mode="human")
    env = gym.make("Mouse1massEnv-v0", render_mode="rgb_array")  # 環境の作成
    # env = gym.make("InvertedPendulum-v4", render_mode="human")

    # # make_vec_envを使用して複数の環境を作成
    # env = make_vec_env(make_env, n_envs=4)

    # video_path = "./"  # 保存先のpath
    # env = RecordVideo(env, video_path, video_length=500)
    # env.model.opt.timestep = 0.01  # タイムステップを設定 RecordVideoする場合はそれ以降に書かないとerrorが出る


    # # 明示的にデフォルトの設定を記述
    # policy_kwargs = dict(
    #     net_arch=[50, 50],
    #     activation_fn=nn.ReLU
    # )

    # model = PPO("MlpPolicy", env, policy_kwargs=policy_kwargs, verbose=0)
    # model = PPO("MlpPolicy", env, verbose=0)

    # # ダミー入力データを作成
    # sample_input = torch.randn((1, env.observation_space.shape[0]))

    # # ポリシーネットワークの出力を計算
    # output = model.policy(sample_input)

    # # ネットワーク構造を可視化
    # dot = make_dot(output, params=dict(model.policy.named_parameters()))
    # dot.render("policy_network", format="png")

    xpos_graph = []
    ypos_graph = []
    # 環境の作成
    env = gym.make("Mouse1massEnv-v0", render_mode="human")
    model = PPO.load(r'c:\Users\atusi\Downloads\saigo3300000steps', env)

    # スタック判定用の変数
    velocity_threshold = 0.08  # 速度がこの値以下ならスタックを疑う
    action_threshold = 0.4  # 大きなアクションを出しているとみなす閾値
    stack_steps = 20  # 何ステップ連続でこの状態ならスタックと判定
    cooldown_steps = 6  # リカバリー後のスタック判定無効期間
    recovery_count = 0

    velocity_history = []
    action_history = []

    # 回避動作用の変数
    avoid_steps = 0  # 回避動作を実行する残りステップ数
    avoid_actions = []  # 回避動作のアクションシーケンス
    cooldown_counter = 0  # クールダウンカウンター

    for i in range(1):
        obs, _ = env.reset()
        
        for j in range(1500):

            if avoid_steps > 0:
                # 回避動作を実行
                # action = avoid_actions.pop(0)
                # avoid_steps -= 1
                apatt=1

            else:
                # クールダウン期間中はスタック判定をしない
                if cooldown_counter > 0:
                    cooldown_counter -= 1
                else:
                    # モデルの推論
                    action, _ = model.predict(obs)

                    # 1ステップ実行
                    obs, reward, terminated, truncated, info = env.step(action)

                    # 速度情報を取得
                    velocity = env.data.sensor("Veloci").data[0]  # 仮定

            #         # 速度履歴を更新
            #         velocity_history.append(velocity)
            #         if len(velocity_history) > stack_steps:
            #             velocity_history.pop(0)

            #         # アクション履歴を更新
            #         action_history.append((abs(action[0]) + abs(action[1])) / 2)
            #         if len(action_history) > stack_steps:
            #             action_history.pop(0)

            #         # スタック判定
            #         if len(velocity_history) == stack_steps and len(action_history) == stack_steps:
            #             max_velocity = max(velocity_history)
            #             max_action = max(action_history)

            #             if max_action > action_threshold and max_velocity < velocity_threshold:
            #                 print("スタック検出！回避動作開始")
            #                 # time.sleep(1)
            #                 # print("velocity_history", velocity_history)
            #                 # print("action_history", action_history)
            #                 recovery_count += 1

            #                 # 回避動作の設定
            #                 avoid_actions = [
            #                     [-0.5, -0.5], [-0.5, -0.5], [-0.5, -0.5],
            #                     [-0.5, -0.5], [-0.5, -0.5], [-0.5, -0.5],
            #                     [0.0, 0.5], [0.0, 0.5], [0.0, 0.5],
            #                     [0.0, 0.5], 
            #                 ]
            #                 avoid_steps = len(avoid_actions)

            #                 # **リカバリー後のクールダウン設定**
            #                 cooldown_counter = cooldown_steps

            #                 # **履歴のクリア**
            #                 velocity_history.clear()
            #                 action_history.clear()

            # # **リカバリー動作中のみマーカーを表示**
            # if avoid_steps > 0:
            #     env.mujoco_renderer.viewer.add_marker(
            #         pos=[env.data.xpos[env.mouse_body_id][0], env.data.xpos[env.mouse_body_id][1], 0], 
            #         size=[0.18, 0.18, 0.18], 
            #         rgba=[1, 0, 0, 0.4], 
            #         type=mujoco.mjtGeom.mjGEOM_BOX, 
            #         label="Recovering..."
            #     )

            # # **マーカーを追加した後に描画**
            # # env.render()

            # # 1ステップ実行
            obs, reward, terminated, truncated, info = env.step(action)
            # blast_flag = 0
            # エピソード完了処理
            if terminated or truncated:
                obs, _ = env.reset()
                # velocity_history.clear()
                # action_history.clear()
                # avoid_steps = 0
                # avoid_actions = []
                # cooldown_counter = 0  # クールダウンもリセット
                # if env.blast_flag == 1:
                #     blast_flag = 1
                break


                
        # #モデルの保存 (1)
        # model.save('M5bala2_standup_1wheel-15zikan(tuikagakusyuuex)')
        # model.save('M5bala2_standup_1wheel-add_posPPO')
        # model.save('nominal_fskip1')
        # print("モデルを保存しました")
            # グラフを作成
        # plt.plot(env.epi_reward_graph)
        # env.stepcount_for_resetgraph = 0 #こうすることで、またステップ前のリセット時にグラフにappendされなくなる
        # # plt.plot(env.deg_difference_graph, label="deg_difference")
        # # plt.plot(env.x_velocity_reward_graph, label="x_velocity_reward")
        # # plt.plot(env.odom_vel_graph, label="x_velocity_reward") 
        # plt.legend()
        # plt.show()
        # model.save('hidarimawari_housyuu_godsight_onlyxy_notight')

    # print(video_path)
    # plt.figure(figsize=(5, 5))  # 図のサイズ
    # plt.title("走行軌跡", fontname="MS Gothic")
    # plt.grid()
    # plt.xlabel("x座標", fontname="MS Gothic")
    # plt.ylabel("y座標", fontname="MS Gothic")
    # 目盛りを手動設定（y軸を例に）
    plt.xticks(np.arange(0, 0.18*16, 0.18))  # 0から0.51まで0.18刻み
    plt.yticks(np.arange(0, 0.18*16, 0.18))  # 0から0.51まで0.18刻み
    # plt.xlim(-0.09, 0.18*9)
    # plt.ylim(-0.09, 0.18*6)
    # 線の開始点と終了点
    lines = [
        ([0.0, 0.0], [0.0, 0.72]),  # 下から上へ（縦線）
        ([0.0, 0.72], [0.72, 0.72]),  # 左から右へ（横線）
        ([0.72, 0.72], [0.72, 0.0]),  # 上から下へ（縦線）
        ([0.72, 0.0], [0.0, 0.0]),  # 左へ
        ([0.18, 0.18], [0.18, 0.54]), # add linepoint  0 till reset
        ([0.18, 0.54], [0.54, 0.54]),  # 下から上へ（縦線）
        ([0.54, 0.54], [0.54, 0.18]),  # 左から右へ（横線）
        ([0.54, 0.18], [0.18, 0.18]),  # 上から下へ（縦線）
    ]
    # ループで描画
    # for x, y in lines:
    #     plt.plot(x, y, color='black', linewidth=4)
    # # plt.legend()
    # plt.plot(xpos_graph, ypos_graph)
    # plt.tight_layout()  # グラフ同士が重ならないように調整
    # # plt.savefig('obal_predict.png', dpi=1000)


    # #前進関係のプロット
    # t = list(range(len(env.reward_graph)))  # 0から始まるインデックスのリストを作成 エピソードの数分報酬（収益）がリストに保存されるので、その数を数えてリストにしただけ　データ数なら自動で出るから必要なかった
    # # plt.plot(t, env.reward_graph, linestyle='solid', label="reward", ) 
    # # plt.legend()
    # plt.show()
    # plt.plot(env.hit_graph, label="hit") 
    # plt.legend()
    # plt.show()
    # # 環境のクローズ
    env.close()

    total = sum(env.hit_graph)
    print("total", total)
    print("走行距離", env.total_distance)
    print("recovery_count", recovery_count)
    total_distance = env.total_distance
    if env.blast_flag == 1:
        print("blast!!!!!!!!!!!!!!!!!!!!!!!!!!!!")
        total_distance = 0
    average.append(total_distance)

    import csv
    # CSVファイルを追記モード（"a"）で開く
    with open("avg_data.csv", mode="a", newline="") as file:
        writer = csv.writer(file)

        # 1回の測定データ
        value = total_distance  # 例：新しいデータ


        # 追記（インデックス付き）
        writer.writerow([value, recovery_count])
print(average)
```
